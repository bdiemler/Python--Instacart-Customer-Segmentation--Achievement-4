```python
# Dropping the unnamed columns created due to Index when exporting
columns_to_drop = ['Unnamed: 0']
```

```python
# Delete the columns and update the DataFrame using 'inplace=True'
df_prods_checked.drop(columns=columns_to_drop, inplace=True)
```

```python
# Check the output df ords
df_prods_checked.head()
```

```python
# Memory error, reducing RAM
# Change types for ords data set
df_ords_combined['order_id']=df_ords_combined['order_id'].astype('int32')
df_ords_combined['user_id'] = df_ords_combined['user_id'].astype('int32')
df_ords_combined['order_number']=df_ords_combined['order_number'].astype('int8')
df_ords_combined['orders_day_of_week']=df_ords_combined['orders_day_of_week'].astype('int8')
df_ords_combined['order_hour_of_day']=df_ords_combined['order_hour_of_day'].astype('int8')
df_ords_combined['days_since_prior_order']=df_ords_combined['days_since_prior_order'].astype('float16')
```

# Merging dataframes

```python
# Create a new dataframe combining both: df_ords_prods_merged
df_ords_prods_merged = df_prods_checked.merge(df_ords_combined, on = 'product_id', indicator = 'Merge_flag')
```

```python
# Check for full match
df_ords_prods_merged.head()
```

```python
# Check for value counts
df_ords_prods_merged['_merge'].value_counts()
```

# Exporting Data
