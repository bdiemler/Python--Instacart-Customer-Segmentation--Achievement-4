### Merging dataframes

Solving memory issues: MemoryError: Unable to allocate 247. MiB for an array with shape (32434212, 1) and data type object

```python
# List current string/object flag columns
object_cols = ['_merge', 'Merge_flag', 'loyalty_flag', 'spender_type', 'order_frequency_flag']
```

```python
for col in object_cols:
    if col in df_ords_prods.columns:
        df_ords_prods[col] = df_ords_prods[col].astype('category')
```

```python
# Create a new dataframe combining both: df_ords_prods_csts_merged
df_ords_prods_csts_merged = df_csts.merge(df_ords_prods, on = 'user_id', indicator = 'Merge_flag49')
```
