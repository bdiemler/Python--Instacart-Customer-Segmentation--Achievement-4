```python
# Item 5 - Finding Missing Values 
df_ords.isnull().sum()
```

```python
# Finding Missing Days in cells
df_nan_days = df_ords[df_ords['days_since_prior_order'].isnull() == True]
```

```python
# Checking dates
df_nan_days
```

Answer item 5: I found 206,209 rows of missing data. When analysing some rows with the null cell, I noticed that all the order numbers = 1. This means that this might not be a dataset mistake, but it is actually related to first orders. As there were no orders before, the # of days prior order will not exist, which returns result null. Thefore, I would not substitute with mean or median, for example, as this would no be accurate and could change the data. 

```python
# Replacing Missing Values with -1 (C:\Users\Aricia\AppData\Local\Temp\ipykernel_46376\159452663.py:2: FutureWarning: A value is trying to be set on a copy of a DataFrame or Series through chained assignment using an inplace method.The behavior will change in pandas 3.0. This inplace method will never work because the intermediate object on which we are setting values always behaves as a copy).
df_ords['days_since_prior_order'].fillna(-1, inplace=True)
```

```python
# Checking replaced values
df_ords.isnull().sum()
```

Answer item 6: As mentioned on previous exercise, the Null cells were related to first orders, which logically did not have a past order. For that reason, I decided to replace values with a special value, that would not compromise the integrity of the data. For that reason, I decided to replace all null values for -1, as it clearly flags the rows as first orders, since days have to be = > 0, without interfering with the actual order already existent.
